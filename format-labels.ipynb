```python
import pandas as pd
```

```python
reviews = pd.read_csv('merged-data/merged-data.csv')
```

```python
label_data = pd.DataFrame(index=reviews['FULL_TEXT'].unique(), columns=reviews['PHRASE'].unique())
```

```python
label_data
```

```python
for review in reviews['FULL_TEXT'].unique():
    for i, j in reviews[reviews['FULL_TEXT'] == review].iterrows():
        label_data.at[review, j['PHRASE']] = 1
```

```python
label_data.fillna(0, inplace=True)
```

```python
label_data.reset_index(inplace=True)
```

```python
label_data.rename(columns={'index': 'FULL_TEXT'}, inplace=True)
```

```python
# dataFrame.to_csv(filepath)
label_data.to_csv('input-data/input-data.csv')
```
